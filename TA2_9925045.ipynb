## outliers

```python
df_1.mean()
```

```python
upper_limit = df_1.mean() + 3 * df_1.std()
upper_limit
```

```python
lower_limit = df_1.mean() - 3 * df_1.std()
lower_limit
```

```python
df_1.columns
```

```python
temp = ['fixed acidity', 'volatile acidity', 'citric acid', 'residual sugar',
       'chlorides', 'free sulfur dioxide', 'total sulfur dioxide', 'density',
       'pH', 'sulphates', 'alcohol']
```

```python
for i in temp:
    lower_limit = df_1[i].mean() - 3 * df_1[i].std()
    upper_limit = df_1[i].mean() + 3 * df_1[i].std()
    df_1 = df_1[(df_1[i] < upper_limit) & (df_1[i] > lower_limit)]
df_1
```

```python

df_2 = df_1[['fixed acidity', 'volatile acidity', 'citric acid', 'residual sugar',
       'chlorides', 'free sulfur dioxide', 'total sulfur dioxide', 'density',
       'pH', 'sulphates', 'alcohol', 'quality']]
df_2
```
